```python
# Import libraries
import fitz

# Extract text from pdf
def extract_text_from_pdf(pdf_path):
    text = ""
    with fitz.open(pdf_path) as doc:
        for page in doc:
            text += page.get_text("text")
    return text

# Import gemini
import google.generativeai as genai
import config

genai.configure(api_key="AIzaSyCSC0LPUznCj0USGxAVXjXT_4vgVqp-ah4")

prompt_template = """
You are an expert IELTS data organizer. 
Below is the extracted text from an IELTS Listening PDF.
Your task:
1. Identify each Part (Part 1, Part 2, Part 3, Part 4).
2. For each Part, separate the text into:
- Question
- Answer
- Transcript

Return the result as **JSON** in the following structure:

{{
"Part 1": {{
    "Question": "...",
    "Answer": "...",
    "Transcript": "..."
}},
"Part 2": {{
    "Question": "...",
    "Answer": "...",
    "Transcript": "..."
}},
"Part 3": {{
    "Question": "...",
    "Answer": "...",
    "Transcript": "..."
}},
"Part 4": {{
    "Question": "...",
    "Answer": "...",
    "Transcript": "..."
}}
}}

If some parts are missing, just leave them empty ("").
Here is the PDF content:
---
{text}
---
"""

# Sent to Gemini
from json import loads

def extract_ielts_parts(text):
    prompt = prompt_template.format(text=text)
    model = genai.GenerativeModel("gemini-2.5-flash")
    response = model.generate_content(prompt)
    try:
        structured = loads(response.text)
        return structured
    except Exception:
        print("Could not parse JSON, here's the raw output:")
        print(response.text)
        return None

pdf_path = "sample_test/sample_1.pdf"
text = extract_text_from_pdf(pdf_path)
structured_data = extract_ielts_parts(text)

if structured_data:
    import json
    print(json.dumps(structured_data, indent=2, ensure_ascii=False))
```
